## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### **Formulated Core Research Hypotheses**

Based on our exploratory data analysis, we will formally test the following three statistical hypotheses:
1.  **Statement 1:** The true population mean closing price of Yes Bank stock during its high-growth phase (2014–2018) is significantly higher than its historical mean from the early phase (2005–2013).
2.  **Statement 2:** The opening pricing feature (`Open`) exhibits a statistically significant linear correlation with the final target variable (`Close`).
3.  **Statement 3:** The distribution of the log-transformed closing stock price matches a normal distribution pattern.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

* **Null Hypothesis ($H_0$):** $\mu_{2014-2018} = \mu_{2005-2013}$ (The mean closing price is identical across both phases).
* **Alternate Hypothesis ($H_1$):** $\mu_{2014-2018} > \mu_{2005-2013}$ (The mean closing price during the growth phase is significantly higher).

#### 2. Perform an appropriate statistical test.

````python
from scipy import stats

early_phase = df[df['Year'] <= 2013]['Close']
growth_phase = df[(df['Year'] >= 2014) & (df['Year'] <= 2018)]['Close']

t_stat, p_val = stats.ttest_ind(growth_phase, early_phase, equal_var=False, alternative='greater')
print(f"Statistical Test Metrics -> T-Statistic: {t_stat:.4f}, P-Value: {p_val:.6e}")
````

##### Which statistical test have you done to obtain P-Value?

We performed a **Welch's Two-Sample Independent t-test** (one-tailed) with unequal variance assumptions to compare the mean closing prices across the two historical phases.

````python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

y_pred_lr = lr_model.predict(X_test)
print("Baseline OLS Linear Regression training process complete.")
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

For predicting the stock's closing price, we have implemented Ordinary Least Squares (OLS) Linear Regression as our baseline model (ML Model - 1). Stock prices naturally follow a continuous numerical trend, making this a classic regression task.
The model creates a linear hyperplane mapping our features (Open, High, Low) directly to the target variable (Close). Since short-term stock prices are highly correlated with their daily opening and intraday price bounds, a linear baseline evaluates structural patterns cleanly. To gauge its performance, we evaluate it on the testing split using standard regression metrics: Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and the R-Squared (‭$R^2$‬) Score to determine the percentage of variance explained.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred_lr)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_lr)

metrics_data = {
    'Evaluation Metric': ['Mean Squared Error (MSE)', 'Root Mean Squared Error (RMSE)', 'R-Squared Score (R2)'],
    'Score Value': [round(mse, 4), round(rmse, 4), round(r2, 4)],
    'Ideal Range': ['Closer to 0 (Lower is better)', 'Closer to 0 (Lower is better)', 'Closer to 1 (Higher is better)']
}
metrics_df = pd.DataFrame(metrics_data)

print("\n========== EVALUATION METRIC SCORE CHART ==========")
print(metrics_df.to_string(index=False))
print("===================================================\n")

plt.figure(figsize=(8, 4))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='Score Value', y='Evaluation Metric', data=metrics_df.iloc[:2], palette='viridis')
plt.title('Error Metrics Comparison (Lower is Better)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Error Value', fontsize=12)
plt.ylabel('Metrics', fontsize=12)
for container in ax.containers:
    ax.bar_label(container, fmt='%.4f', padding=5, fontweight='bold')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 2))
ax_r2 = sns.barplot(x=[r2], y=['R-Squared Score (R2)'], color='#2ecc71')
plt.title('Model Variance Explanation / Accuracy (Higher is Better)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Score Value (Scale 0 to 1)', fontsize=12)
plt.xlim(0, 1.1)
for container in ax_r2.containers:
    ax_r2.bar_label(container, fmt='%.4f', padding=5, fontweight='bold')
plt.tight_layout()
plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

For a simple standard OLS Linear Regression model, there are no primary structural hyperparameters to tune (unlike deep learning weights or tree depths in random forests). The coefficients are mathematically solved using the normal equation or singular value decomposition to minimize the residual sum of squares directly.
To test stability and check for overfitting without tuning knobs, we implement a K-Fold Cross-Validation structure. This splits our scaled feature data multiple times to ensure that our ‭$R^2$‬ accuracy is consistent across different chunks of the dataset and not biased by a single random split.

````python
from sklearn.model_selection import cross_val_score

# Performing a 5-Fold Cross Validation on our baseline model to check structural stability
cv_scores = cross_val_score(lr_model, X_scaled, y, cv=5, scoring='r2')

print("========== CROSS-VALIDATION RESULTS (5-FOLDS) ==========")
for idx, score in enumerate(cv_scores, 1):
    print(f"Fold {idx} R2 Score: {score:.4f}")

print("--------------------------------------------------------")
print(f"Mean Cross-Validation R2 Score: {np.mean(cv_scores):.4f}")
print(f"Standard Deviation of CV Scores: {np.std(cv_scores):.4f}")
print("========================================================")
````

##### Which hyperparameter optimization technique have you used and why?

None (K-Fold Cross-Validation evaluated instead). Since baseline OLS Linear Regression relies on deterministic algebraic equations to fit coefficients, it lacks regularized hyperparameters (like alpha or penalty parameters found in Ridge/Lasso). Therefore, optimization techniques like GridSearchCV or RandomSearchCV do not apply here. Instead, we used 5-Fold Cross-Validation to assess structural stability and guard against performance overfitting across the dataset.

### ML Model - 2

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

For ML Model - 2, we are implementing Ridge Regression (L2 Regularization). Since stock market features like Open, High, and Low are highly correlated with each other (multicollinearity), standard linear regression can sometimes have high variance. Ridge Regression solves this by adding a penalty equivalent to the square of the magnitude of coefficients (‭$L2$‬ penalty). This keeps the weights stable, prevents overfitting, and ensures the model performs reliably on real-world, unseen financial data.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score

# 1. Initialize and Train Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)

# 2. Predict on Test Set
y_pred_ridge = ridge_model.predict(X_test)

# 3. Calculate Performance Metrics
mse_r = mean_squared_error(y_test, y_pred_ridge)
rmse_r = np.sqrt(mse_r)
r2_r = r2_score(y_test, y_pred_ridge)

# 4. Create Metric Score Table
metrics_ridge_df = pd.DataFrame({
    'Evaluation Metric': ['Mean Squared Error (MSE)', 'Root Mean Squared Error (RMSE)', 'R-Squared Score (R2)'],
    'Score Value': [round(mse_r, 4), round(rmse_r, 4), round(r2_r, 4)]
})

print("\n========== ML MODEL - 2: RIDGE REGRESSION SCORE CHART ==========")
print(metrics_ridge_df.to_string(index=False))
print("=================================================================\n")

# 5. Visualizing the scores
fig, ax = plt.subplots(1, 2, figsize=(14, 4))
sns.barplot(x='Score Value', y='Evaluation Metric', data=metrics_ridge_df.iloc[:2], ax=ax[0], palette='mako')
ax[0].set_title('Ridge Regression Error Metrics (Lower is Better)', fontweight='bold')
for c in ax[0].containers: ax[0].bar_label(c, fmt='%.4f', padding=5)

sns.barplot(x=[r2_r], y=['R-Squared Score (R2)'], ax=ax[1], color='#3498db')
ax[1].set_title('Ridge Regression Accuracy / R2 (Higher is Better)', fontweight='bold')
ax[1].set_xlim(0, 1.1)
for c in ax[1].containers: ax[1].bar_label(c, fmt='%.4f', padding=5)
plt.tight_layout()
plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

To find the absolute top-class configuration for Ridge Regression, we perform hyperparameter tuning on the regularization strength parameter alpha (which controls the penalty complexity). We use GridSearchCV with 5-Fold Cross-Validation to evaluate an array of alpha values ranging from very small to large fractions, ensuring we hit the sweet spot between bias and variance.

````python
from sklearn.model_selection import GridSearchCV

# Defining parameter grid for Alpha
param_grid = {'alpha': [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]}

# GridSearch Setup
grid_ridge = GridSearchCV(Ridge(), param_grid, cv=5, scoring='r2')
grid_ridge.fit(X_train, y_train)

# Best Model Extraction
best_ridge = grid_ridge.best_estimator_
y_pred_tuned_ridge = best_ridge.predict(X_test)

print(f"Optimal Alpha Selected by GridSearchCV: {grid_ridge.best_params_['alpha']}")
print(f"Optimized Model Top R2 Score: {r2_score(y_test, y_pred_tuned_ridge):.4f}")
````

##### Which hyperparameter optimization technique have you used and why?

We utilized GridSearchCV (Grid Search Cross-Validation). Since Ridge Regression has a single major continuous parameter (alpha), a grid search allows us to comprehensively scan across logarithmic steps to pinpoint the precise mathematical weight that penalizes collinearity effectively without underfitting the trend lines.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes, marginal improvement and cleaner convergence. While the baseline OLS model was already strong, tuning the alpha parameter reduced the impact of minor intraday market noise. The final updated Evaluation metric Score Chart shows stable metrics with reduced multi-collinearity stress on the weights.
