````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# File load kar rahe hain
df = pd.read_csv('data_YesBank_StockPrices.csv')
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')

print("Data loaded successfully! Total rows:", df.shape[0])
print(df.head())
````

````python
# Trend over time graph
plt.figure(figsize=(10, 4))
plt.plot(df['Date'], df['Close'], color='green', linewidth=2)
plt.title('Yes Bank Closing Price Trend')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()

# Heatmap graph
plt.figure(figsize=(6, 4))
sns.heatmap(df[['Open', 'High', 'Low', 'Close']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
````
